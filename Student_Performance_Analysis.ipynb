```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data collection
```

```python
data = pd.read_csv('/kaggle/input/students-performance-in-exams/StudentsPerformance.csv')
```

```python
student = data.drop(['race/ethnicity','parental level of education'],axis=1)
```

```python
student.head()
```

```python
#relationship analysis
```

```python
scores = student[['math score', 'reading score', 'writing score']]
correlation = scores.corr()

sns.heatmap(correlation, annot=True, cmap="coolwarm")
```

```python
sns.pairplot(student)
```

```python
gender_counts = student['gender'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['pink', 'skyblue'])
plt.title('Distribution of Students by Gender')
plt.ylabel('') # Hide the default 'gender' label
plt.show()
```

```python
plt.scatter(x='math score',y='writing score',data=student,color = 'pink')
plt.title('student performance')
plt.show
```

```python
sns.displot(student['math score'])
```

```python
sns.displot(student['writing score'])
```

```python
sns.distplot(student['reading score'], bins = 20)
```

```python
sns.catplot(x = 'math score',kind = 'box', data = student)
```

```python
sns.catplot(x = 'reading score', kind = 'box', data = student)
```

```python
sns.catplot(x = 'writing score', kind = 'box', data = student)
```

```python
sns.scatterplot(data=student,x='math score',y='reading score', hue='gender')
plt.xlabel('math score')
plt.ylabel('reading score')
plt.title('math vs reading')
```
